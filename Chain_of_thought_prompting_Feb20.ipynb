```python
def solve_math_problem(problem):
    prompt = f"""Solve the following math word problem step by step:

Problem: {problem}

Please follow these steps:
1. Identify the important information given in the problem.
2. Determine what the question is asking.
3. Plan the steps to solve the problem.
4. Execute each step, showing your work.
5. Provide the final answer.

Let's approach this step-by-step:"""

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a helpful math tutor that solves word problems step by step."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=500,
        temperature=0.7
    )

    return response.choices[0].message.content
```

```python
# Example usage
problem = "A store is having a 20% off sale. If a shirt originally costs $45, how much will it cost after the discount?"
solution = solve_math_problem(problem)
print(solution)
```

```python
def make_decision(scenario):
    prompt = f"""Consider the following scenario and make a decision step by step:

Scenario: {scenario}

Please follow these steps:
1. Identify the key factors and stakeholders involved.
2. List the potential options or courses of action.
3. Evaluate the pros and cons of each option.
4. Consider any ethical implications or long-term consequences.
5. Make a final recommendation and justify your decision.

Let's approach this step-by-step:"""

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a thoughtful advisor who carefully considers complex scenarios."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=500,
        temperature=0.7
    )

    return response.choices[0].message.content
```

```python
# Example usage
scenario = "A small tech company has developed a groundbreaking AI algorithm. They've received offers from both a large tech giant wanting to acquire the technology exclusively, and a non-profit organization proposing to make it open-source for the benefit of smaller businesses and researchers. The company needs to decide which path to take."

decision = make_decision(scenario)
print(decision)
```

```python
from openai import OpenAI

class FlexibleCoTSolver:
    def __init__(self, api_key):
        self.client = OpenAI(api_key=api_key)

    def classify_question(self, question):
        classification_prompt = f"""
Classify the following question into one of these categories: math, decision, explanation, or general.
Only respond with the category name, nothing else.

Question: {question}

Category:"""

        response = self.client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a question classifier. Categorize the given question accurately."},
                {"role": "user", "content": classification_prompt}
            ],
            max_tokens=10,
            temperature=0.3
        )

        return response.choices[0].message.content.strip().lower()

    def generate_prompt(self, question_type, question):
        base_prompt = f"Question: {question}\n\nLet's approach this step-by-step:"

        prompts = {
            "math": """
1. Identify the important information given in the problem.
2. Determine what the question is asking.
3. Plan the steps to solve the problem.
4. Execute each step, showing your work.
5. Provide the final answer.""",
            "decision": """
1. Identify the key factors and stakeholders involved.
2. List the potential options or courses of action.
3. Evaluate the pros and cons of each option.
4. Consider any ethical implications or long-term consequences.
5. Make a final recommendation and justify your decision.""",
            "explanation": """
1. Define any key terms or concepts.
2. Break down the topic into its core components.
3. Explain each component in detail.
4. Provide examples or analogies to illustrate the concept.
5. Summarize the main points and their significance.""",
            "general": """
1. Analyze the key elements of the question.
2. Break down the problem into smaller, manageable parts.
3. Address each part systematically.
4. Synthesize the information to form a comprehensive answer.
5. Conclude with a summary of the main points."""
        }

        return base_prompt + prompts.get(question_type, prompts["general"])

    def solve_problem(self, question):
        question_type = self.classify_question(question)
        prompt = self.generate_prompt(question_type, question)

        response = self.client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a versatile problem-solver capable of addressing various types of questions step-by-step."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=500,
            temperature=0.7
        )

        return f"Question Type: {question_type.capitalize()}\n\nAnswer: {response.choices[0].message.content}"
```
