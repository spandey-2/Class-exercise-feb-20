```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cpu",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

# Create a pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=False,
)
```

```python
# Prompt
messages = [
    {"role": "user", "content": "Create a funny joke about indian airlines."}
]

# Generate the output
output = pipe(messages)
print(output[0]["generated_text"])
```

```python
output
```

```python
# Apply prompt template
prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False)
print(prompt)
```

```python
# Using a high temperature
output = pipe(messages, do_sample=True, temperature=1)
print(output[0]["generated_text"])
```

```python
# Using a high top_p
output = pipe(messages, do_sample=True, top_p=1)
print(output[0]["generated_text"])
```

```python
output = pipe(prompt, temperature=1)

# Print the generated text
print(output[0]["generated_text"])
```

# **Advanced Prompt Engineering**


## Complex Prompt

```python
text = """In a recent study on airlines research, researchers examined advanced methodologies to optimize airline operations and improve overall efficiency. The study delved into critical aspects such as fuel efficiency, predictive maintenance, route optimization, and enhanced customer service. Advanced data analytics and machine learning techniques were leveraged to compare traditional methods with modern approaches.

Key findings include:
- Integration of simulation models to predict flight delays and optimize scheduling using real-time data from weather, air traffic, and airspace management systems.
- Implementation of predictive maintenance algorithms that detect potential failures before they occur, thereby reducing downtime and enhancing safety.
- Adoption of innovative strategies for sustainability, such as optimized flight path planning and the use of fuel-efficient aircraft to reduce carbon emissions.

Overall, the research provides a comprehensive roadmap for future innovations in the aviation sector, encouraging collaboration among airlines, technology providers, and regulatory bodies to tackle emerging challenges in the industry."""

# Updated prompt components
persona = "You are an expert in aviation research and advanced data analytics. You excel at breaking down complex studies into digestible summaries.\n"
instruction = "Summarize the key findings of the study provided.\n"
context = "Your summary should extract the most crucial points that can help industry professionals quickly understand the essential insights of the study.\n"
data_format = "Create a bullet-point summary that outlines the methods and findings, followed by a concise paragraph that encapsulates the main results.\n"
audience = "The summary is designed for busy industry professionals and researchers in the aviation sector.\n"
tone = "The tone should be professional and clear.\n"
data = f"Text to summarize: {text}"

# The full prompt
query = persona + instruction + context + data_format + audience + tone + data
```

```python
query
```

```python
messages = [
    {"role": "user", "content": query}
]
print(tokenizer.apply_chat_template(messages, tokenize=False))
```

```python
# Generate the output
outputs = pipe(messages)
print(outputs[0]["generated_text"])
```

## In-Context Learning: Providing Examples

```python
# Use a single example of using the made-up word in a sentence
one_shot_prompt = [
    {
        "role": "user",
        "content": "A 'Gigamuru' is a type of Japanese musical instrument. An example of a sentence that uses the word Gigamuru is:"
    },
    {
        "role": "assistant",
        "content": "I have a Gigamuru that my uncle gave me as a gift. I love to play it at home."
    },
    {
        "role": "user",
        "content": "To 'screeg' something is to swing a sword at it. An example of a sentence that uses the word screeg is:"
    }
]
print(tokenizer.apply_chat_template(one_shot_prompt, tokenize=False))
```

```python
# Generate the output
outputs = pipe(one_shot_prompt)
print(outputs[0]["generated_text"])
```

## Chain Prompting: Breaking up the Problem


```python
# Create name and slogan for a product
product_prompt = [
    {"role": "user", "content": "Create a name and slogan for a chatbot that leverages LLMs."}
]
outputs = pipe(product_prompt)
product_description = outputs[0]["generated_text"]
print(product_description)
```

```python
# Based on a name and slogan for a product, generate a sales pitch
sales_prompt = [
    {"role": "user", "content": f"Generate a very short sales pitch for the following product: '{product_description}'"}
]
outputs = pipe(sales_prompt)
sales_pitch = outputs[0]["generated_text"]
print(sales_pitch)
```

# **Reasoning with Generative Models**


## Chain-of-Thought: Think Before Answering


```python
# Answering without explicit reasoning
standard_prompt = [
    {"role": "user", "content": "Roger has 5 tennis balls. He buys 2 more cans of tennis balls. Each can has 3 tennis balls. How many tennis balls does he have now?"},
    {"role": "assistant", "content": "11"},
    {"role": "user", "content": "An airline's aircraft is scheduled to depart with a minimum fuel load of 10,000 liters. However, because fuel is significantly cheaper at the departure airport, the airline decides to engage in fuel tankering. It loads 2 extra fuel tanks, each holding 3,000 liters. How many liters of fuel does the aircraft have on board now?"}
]

# Run generative model
outputs = pipe(standard_prompt)
print(outputs[0]["generated_text"])
```

```python
# Answering with chain-of-thought
cot_prompt = [
    {"role": "user", "content": "Roger has 5 tennis balls. He buys 2 more cans of tennis balls. Each can has 3 tennis balls. How many tennis balls does he have now?"},
    {"role": "assistant", "content": "Roger started with 5 balls. 2 cans of 3 tennis balls each is 6 tennis balls. 5 + 6 = 11. The answer is 11."},
    {"role": "user", "content": "An airline's aircraft is scheduled to depart with a minimum fuel load of 10,000 liters. However, because fuel is significantly cheaper at the departure airport, the airline decides to engage in fuel tankering. It loads 2 extra fuel tanks, each holding 3,000 liters. How many liters of fuel does the aircraft have on board now?"}
]

# Generate the output
outputs = pipe(cot_prompt)
print(outputs[0]["generated_text"])
```

## Zero-shot Chain-of-Thought


```python
# Zero-shot Chain-of-Thought
zeroshot_cot_prompt = [
    {"role": "user", "content": "An airline's aircraft is scheduled to depart with a minimum fuel load of 10,000 liters. However, because fuel is significantly cheaper at the departure airport, the airline decides to engage in fuel tankering. It loads 2 extra fuel tanks, each holding 3,000 liters. How many liters of fuel does the aircraft have on board now? Let's think step-by-step."}
]

# Generate the output
outputs = pipe(zeroshot_cot_prompt)
print(outputs[0]["generated_text"])
```

## Tree-of-Thought: Exploring Intermediate Steps


```python
# Zero-shot Chain-of-Thought
zeroshot_tot_prompt = [
    {"role": "user", "content": "Imagine three different experts are answering this question. All experts will write down 1 step of their thinking, then share it with the group. Then all experts will go on to the next step, etc. If any expert realises they're wrong at any point then they leave. The question is 'An airline's aircraft is scheduled to depart with a minimum fuel load of 10,000 liters. However, because fuel is significantly cheaper at the departure airport, the airline decides to engage in fuel tankering. It loads 2 extra fuel tanks, each holding 3,000 liters. How many liters of fuel does the aircraft have on board now? Let's think step-by-step.' Make sure to discuss the results."}
]
```

```python
# Generate the output
outputs = pipe(zeroshot_tot_prompt)
print(outputs[0]["generated_text"])
```
